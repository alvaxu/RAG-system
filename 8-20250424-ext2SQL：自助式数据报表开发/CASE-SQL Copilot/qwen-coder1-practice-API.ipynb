```python
import os
from openai import OpenAI  # 使用兼容OpenAI接口的SDK

# 1. 初始化客户端（需替换为你的百炼API Key）
MODEL = "qwen2.5-coder-7b-instruct"  # 百炼平台上的模型名称
client = OpenAI(
    api_key=os.getenv("DASHSCOPE_API_KEY"),  # 建议通过环境变量配置API Key
    base_url="https://dashscope.aliyuncs.com/compatible-mode/v1"  # 百炼API端点
)
# 2.定义response
def get_response(messages):
    response = client.chat.completions.create(  # 注意接口路径变化
        model= MODEL,
        messages= messages,
        max_tokens=512,       # 生成最大token数
        temperature=0.7  # 控制回答的随机性（0-1）
    )
    return response.choices[0].message.content

# 3. 构造对话请求（与本地调用格式兼容）
prompt = "帮我编写用户登录的测试用例"
messages = [
    {"role": "system", "content": "You are a helpful assistant."},
    {"role": "user", "content": prompt}
]

# 4. 输出模型响应
response = get_response(messages)
print(response)
```
